```python
import psycopg2
import pandas as pd
```

```python
password = input("please input your pw: ")
```

```python
# Establish database connection
conn = psycopg2.connect(
    host="localhost",
    port="5432",
    database="PendaHealth",
    user="postgres",
    password=password)
conn
```

```python
# Create a cursor
cursor = conn.cursor()
```

```python
# QUERIES
diagnosis = 'SELECT * FROM Diagnosis_tbl'
visit = 'SELECT * FROM Visit_tbl'
invoice = 'SELECT * FROM Invoice_tbl'
```

# Executing SQL query to pull visit table from the DB

```python
cursor.execute(visit)

#Fetching invoice data
visit_data = cursor.fetchall()

#Getting the pulled Visit data into a DF
visit = pd.DataFrame(visit_data,columns=[desc[0] for desc in cursor.description])
visit
```

# Executing SQL query to pull invoice table from the DB

```python
cursor.execute(invoice)

# Fetching invoice data
invoice_data = cursor.fetchall()

# Getting the pulled Invoice_data to a DF
invoice = pd.DataFrame(invoice_data,columns=[desc[0] for desc in cursor.description])
```

# Execute SQL query to get diagnosis table

```python
cursor.execute(diagnosis)
# Fetch data
diagnosis_data = cursor.fetchall()
```

```python
#Diagnosis_data 
diagnosis = pd.DataFrame(diagnosis_data,columns=[desc[0] for desc in cursor.description])
diagnosis
```

```python
#Close connection
cursor.close()
conn.close()
```
